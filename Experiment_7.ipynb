```python
# Print numbers from 1 to N
n = int(input("Enter N: "))
i = 1
while i <= n:
    print(i)
    i += 1
```

```python
# Factorial of a number
n= int(input("Enter a number:"))
fact = 1
while n>0:
  fact *= n
  n -= 1
print("Factorial: ", fact)
```

```python
# Fibonacci series using while loop
n = int(input("Enter the number of terms: "))
a = 0
b = 1
i = 1
while i <= n:
  print(a, end=" ")
  c = a + b
  a = b
  b = c
  i += 1
```

```python
# Adding limit to Fibonacci series
limit = int(input("Enter the limit: "))
a = 0
b = 1
while a <= limit:
  print(a, end=" ")
  a, b = b, a+b
```

```python
# Reverse a number
num =int(input("Enter a number: "))
rev = 0
while num > 0 :
         digit = num % 10
         rev = rev * 10 + digit
         num //= 10
print("Reversed number: ", rev)
```

```python
# Check Palindrome number
num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
    rev = rev * 10 + num % 10
    num //= 10
if temp == rev:
  print("The number is Palindrome")
else:
  print("The number is not Palindrome")
```

```python
# Checking Palidrome for string
s = input("Enter a string: ")
i, j = 0, len(s) - 1
is_palindrome = True
while i < j:
    if s[i] != s[j]:
        is_palindrome = False
        break
    i += 1
    j -= 1
if is_palindrome:
    print("The string is a palindrome.")
else:
    print("The string is not a palindrome.")
```

```python
# Another method for checking Palindrome
st = input("Enter a string: ")
reverse = st[::-1]
if st == reverse:
  print("The string is Palindrome")
else:
  print("The string is not Palindrome")
```

```python
# Count digits in a number
num = int(input("Enter a number: "))
count = 0
while num > 0:
    num //= 10
    count += 1
print("Total digits are:", count)
```

```python
# Exit the loop when i is 3
i = 1
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
# Search an element in a list
nums = [10, 20,30,40,50]
key = int(input("Enter an element to search in the list: "))
i = 0
while i < len(nums):
    if nums[i] == key:
      print("Element found at index", i+1)
      break
    i += 1
else:
  print("Element not found in the list")
```
